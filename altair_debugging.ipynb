### 차트 만들기 시도 (Try Making a Chart)

## 디스플레이 문제 해결 (Display Troubleshooting)

차트가 보이지 않는 문제가 발생하면, 위의 [디버깅 지침](#Installation)에 따라 설정이 올바른지 확인하세요. 여전히 문제가 해결되지 않는다면, [Altair 문서의 디스플레이 문제 해결에 관한 지침](https://iliatimofeev.github.io/altair-viz.github.io/user_guide/troubleshooting.html)을 따르세요.

### 존재하지 않는 필드 (Non Existent Fields)

흔히 발생하는 오류 중 하나는 [실제로 존재하지 않는 필드를 실수로 사용](https://iliatimofeev.github.io/altair-viz.github.io/user_guide/troubleshooting.html#plot-displays-but-the-content-is-empty)하는 것입니다.

````python
import pandas as pd

df = pd.DataFrame({'x': [1, 2, 3],
                     'y': [3, 1, 4]})

alt.Chart(df).mark_point().encode(
    x='x:Q',
    y='y:Q',
    color='color:Q'  # <-- this field does not exist in the data!
)
````

파일의 철자를 확인하고 데이터 소스를 출력하여 데이터와 필드가 존재하는지 확인하세요. 예를 들어, 여기서는 `color`가 유효한 필드가 아님을 알 수 있습니다.

````python
df.head()
````

## 유효하지 않은 사양 (Invalid Specifications)

또 다른 흔한 문제는 유효하지 않은 사양을 만들어 오류가 발생하는 것입니다.

### 유효하지 않은 속성 (Invalid Properties)

Altair는 `SchemaValidationError` 또는 `ValueError`를 표시할 수 있습니다. 오류 메시지를 주의 깊게 읽어보세요. 보통 무엇이 잘못되었는지 알려줄 것입니다.

예를 들어, 마크 유형을 잊어버린 경우 다음과 같은 `SchemaValidationError`를 보게 됩니다.

````python
alt.Chart(cars).encode(
    y='Horsepower'
)
````

또는 존재하지 않는 채널을 사용하면 `ValueError`가 발생합니다.

````python
alt.Chart(cars).mark_point().encode(
    z='Horsepower'
)
````

## 속성이 무시됨 (Properties are Being Ignored)

Altair는 사용자가 지정한 속성을 무시할 수 있습니다. 아래 차트에서는 `mark_text`와만 호환되는 `text` 채널을 사용하고 있습니다. 노트북에서는 이에 대한 오류나 경고가 표시되지 않습니다. 그러나 기초가 되는 Vega-Lite 라이브러리는 브라우저 콘솔에 경고를 표시합니다. Mac에서는 <kbd>Alt</kbd>+<kbd>Cmd</kbd>+<kbd>I</kbd>, Windows 및 Linux에서는 <kbd>Alt</kbd>+<kbd>Ctrl</kbd>+<kbd>I</kbd>를 눌러 개발자 도구를 열고 `Console` 탭을 클릭하세요. 아래 셀의 예제를 실행하면 다음과 같은 경고가 표시됩니다.

```
WARN text dropped as it is incompatible with "bar".
```

````python
alt.Chart(cars).mark_bar().encode(
    y='mean(Horsepower)',
    text='mean(Acceleration)'
)
````

Vega-Lite와 관련된 문제를 디버깅해야 하는 경우, 차트 하단의 "Open in Vega Editor" 링크를 클릭하거나 차트 오른쪽 상단의 작업 메뉴(클릭하여 열기)에서 [Vega Editor](https://vega.github.io/editor/)로 차트를 열 수 있습니다. Vega Editor는 추가적인 디버깅 기능을 제공하지만, 파이썬의 Altair 대신 Vega-Lite JSON을 작성하게 됩니다.

**참고**: Vega Editor는 최신 버전의 Vega-Lite를 사용할 수 있으므로 동작이 다를 수 있습니다.
